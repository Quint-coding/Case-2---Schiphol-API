```python

def callFlightsAPI():
    url = 'https://api.schiphol.nl/public-flights/flights?includedelays=false&page=0&sort=%2BscheduleTime'  # Replace with the actual API URL
    headers = {
        "Authorization": "Bearer YOUR_ACCESS_TOKEN",  # Replace with actual authentication if needed
        "Accept": "application/json"
    }

    try:
        response = requests.get(url, headers=headers)  # Make the API call
        response.raise_for_status()  # Handle HTTP errors (4xx and 5xx)
    except requests.exceptions.RequestException as error:
        print(f"Error: {error}")
        sys.exit()

    # Check if response is OK
    if response.status_code == 200:
        flights_data = response.json()  # Convert response to dictionary

        # Verify if "flights" key exists in JSON response
        if "flights" in flights_data:
            flights_list = flights_data["flights"]  # Extract flights list

            # Normalize nested JSON data into a DataFrame
            df = pd.json_normalize(flights_list, sep='_')

            # Display DataFrame
            print(df)

            # Optionally, save to a CSV file
            df.to_csv("flights_data.csv", index=False)
        else:
            print("Unexpected response structure: 'flights' key not found.")
            print(flights_data)
    else:
        print(f"Oops, something went wrong!\nHTTP response code: {response.status_code}\n{response.text}")

# Call the function
callFlightsAPI()
```

```python
import requests
import pandas as pd
import time

url_base = "https://api.schiphol.nl/public-flights/flights"
headers = {
    'accept': 'application/json',
    'resourceversion': 'v4',
    'app_id': 'aaeab06e',
    'app_key': '346684b4f4090227ec9abe2d93526665'
}


all_flights = []  # List to store all pages of flight data
max_pages = 100  # Limit the number of pages to 2

for page in range(max_pages):
    url = f"{url_base}?includedelays=false&page={page}&sort=%2BscheduleTime"

    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 429:  # Handle rate-limiting
            print("Rate limit reached. Waiting before retrying...")
            time.sleep(5)  # Wait 5 seconds before retrying
            response = requests.get(url, headers=headers)  # Retry request

        response.raise_for_status()  # Raise error for bad responses (4xx, 5xx)
    except requests.exceptions.RequestException as error:
        print(f"Error fetching data: {error}")
        break

    data = response.json()

    if "flights" not in data or not data["flights"]:
        print("No more flights found. Stopping pagination.")
        break  # Exit loop when no more data

    all_flights.extend(data["flights"])  # Append flights from current page
    time.sleep(1)  # Pause between requests to avoid hitting rate limits

# Convert list of flights to DataFrame
if all_flights:
    df = pd.json_normalize(all_flights, sep='_')
    print(df)  # Display the full DataFrame
    df.to_csv("flights_data.csv", index=False)  # Optionally save to CSV
else:
    print("No flight data retrieved.")

df = pd.DataFrame(all_flights)  # Convert to DataFrame
df

```
