```python
# Data wordt op verschillende pagina's opgeslagen, door de URL te splitsen kunnen we door de pagina's lopen
url_base = "https://api.schiphol.nl/public-flights/flights"

headers = {
    'accept': 'application/json',
    'resourceversion': 'v4',
    'app_id': 'aaeab06e',
    'app_key': '346684b4f4090227ec9abe2d93526665'
}

max_pages = 4  # Limiet op het aantal pagina's dat doorlopen wordt

@st.cache_data(ttl=3600)  # Cache for 1 hour to avoid repeated requests
def fetch_flight_data(url_base=url_base, max_pages=max_pages, headers=headers):    
    all_flights = []  # Lijst om de data van alle pagina's op te slaan

    for page in range(max_pages):   # Via een for-loop de URL aanpassen
        url = f"{url_base}?includedelays=false&page={page}&sort=%2BscheduleTime"    # de URL aanpassen naar de pagina's
        response = requests.get(url, headers=headers)   # Response defineren via .get() request
        data = response.json()  # Response inlezen met .json()
        all_flights.extend(data["flights"])  # De all_flights dataframe aanvullen met de waardes van de pagina's 
        time.sleep(1)  # korte pauze tussen .get() requests om een rate limit te voorkomen

    df = pd.DataFrame(all_flights)  # nadat alle pagina's zijn ingelezen 

df_cache = fetch_flight_data()

df.head()
```

### Exploratory Data Analysis

```python
df.info()   #opvallend dat de tijd kolommen niet als Datetime staan
```

```python
# Kolommen met meerdere waardes splitsen 

## df['destination'] = df['route'].apply(lambda x: ', '.join(x['destinations']) if 'destinations' in x else None) 
    # deze code geeft alle bestemmingen

# De route kolom splitsen in bestemming, eu en visa 
df['destination'] = df['route'].apply(lambda x: x.get('destinations', [None])[0])   # [0] gebruiken om alleen de eerste waarde uit de lijst te krijgen
df['eu'] = df['route'].apply(lambda x: x.get('eu', [None]))
df['visa'] = df['route'].apply(lambda x: x.get('visa', [None]))

df['vlucht_status'] = df['publicFlightState'].apply(lambda x: x.get('flightStates', [None])[0])    # [0] gebruiken om alleen de eerste waarde uit de lijst te krijgen

df["baggage_belt"] = df["baggageClaim"].apply(
    lambda x: int(x["belts"][0]) if isinstance(x, dict) and "belts" in x and isinstance(x["belts"], list) and x["belts"] else None
).fillna(0).astype(int)     # krijgt de belt nummer uit de dictionairy en maakt daar een integer van

# De aircraftType kolom splitsen in iataMain en iataSub
df['iataMain'] = df['aircraftType'].apply(lambda x: x.get('iataMain', [None]))
df['iataSub'] = df['aircraftType'].apply(lambda x: x.get('iataSub', [None]))

# tijd data opslaan als Datetime datatype
df['actualLandingTime'] = pd.to_datetime(df['actualLandingTime'])

# controle op de nieuwe kollomen
df.head()
```
